```python
%%bash

find $HOME/Programs/tf_basilisk/Benchs -name 'count.csv' > /tmp/csvs.txt
cat /tmp/csvs.txt
# cat /home/guilhermel/Programs/tf_basilisk/Benchs/Misc/richards_benchmark/count.csv
```

```python
dfs = []
with open('/tmp/csvs.txt') as f:
    for line in f:
        df_aux = pd.read_csv(line.strip())
        line_aux = line.strip()\
                        .replace('/home/guilhermel/Programs/tf_basilisk/Benchs/', '')\
                        .replace('/count.csv', '')\
                        .replace('/rbc', '')
        df_aux['benchmark'] = line_aux
        dfs.append(df_aux)
```

```python
df = pd.concat(dfs).set_index('benchmark')
```
